```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('incorrect.csv')
```

```python
import glob

labels = ["i", "ii", "iii", "iv", "v", "vi", "vii", "viii", "ix", "x"]
print("====Training dataset stats====")
for label in labels:
    print(label + "\t" + str(len(glob.glob("data_11-whitespace-aug/train/" + label + "/*png"))))
    

print("====Validation dataset stats====")
for label in labels:
    print(label + "\t" + str(len(glob.glob("data_11-whitespace-aug/val/" + label + "/*png"))))
```

```python
import cv2
from math import ceil

def plot_category(category):
    mini_df = df[df['category'] == category]
    print(mini_df.value_counts('prediction'))
    mini_df = mini_df.sort_values('prediction')

    N = len(mini_df)
    print(N)
    if N > 6:
        div = 6
    else:
        div = 1
    fig, ax = plt.subplots(ceil(N/div),div, figsize=(12,8))
    j = 0
    for i, entry in mini_df.iterrows():
        item = entry['file']
        pred = entry['prediction']
        img = cv2.imread(item)
        if div > 1:
            ax[j//div, j%div].imshow(img)
            ax[j//div, j%div].set_title("prediction: " + pred + " ")
        else:
            ax[j].imshow(img)
            ax[j].set_title("prediction: " + pred + " ")
        j = j+ 1
        if j > N:
            break

    fig.tight_layout()

plot_category('i')
```
